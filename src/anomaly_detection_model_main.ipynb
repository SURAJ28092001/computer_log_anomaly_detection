```python
import pandas as pd
```

```python
import csv
import json
```

```python
with open("D://kaggle/data/train.json") as json_file:
    data = json.load(json_file)
```

```python
len(data.keys())
```

```python
df = pd.json_normalize(data)
df
```

```python
df = df.transpose()
```

```python
df
```

```python
c = 0
df_new = {'0':[],'1':[]}
for i, index in df.iterrows():
    str1 = i.split()[5:]
    str1 = ' '.join(str1)
    df_new['0'].append(str1)
    df_new['1'].append(index[0])
    
df_new = pd.DataFrame(df_new)
df_new
```

```python
import numpy as np
```

```python
df_new['1'] = df_new['1'].apply(lambda x : 0 if x == 'normal' else 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_new['0'],df_new['1'],test_size=0.2,random_state=20)
```

```python
X = np.array_split(x_train,10000)
Y = np.array_split(y_train,10000)
```

```python
from sklearn.naive_bayes import BernoulliNB
model = BernoulliNB()
```

```python
X_test = np.array_split(x_test,10000)
Y_test = np.array_split(y_test,10000)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
vectorizer = TfidfVectorizer()
vectorizer.fit(x_train)
for i in range(0,10000):
    x = X[i]
    y = Y[i]
    x = vectorizer.transform(x)
    x = x.toarray()
    model.partial_fit(x,y,classes=[0,1])
    print(model.score(x,y))
```

```python
backup = model
```

```python
x = x_test[0:2000]
y = y_test[0:2000]
```

```python
x = vectorizer.transform(x)
x = x.toarray()
```

```python
y_predict = model.predict(x)
print(accuracy_score(y,y_predict))
```
