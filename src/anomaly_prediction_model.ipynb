```python
import pickle
with open('anomaly_detection_model.pkl','rb') as f:
    model = pickle.load(f)
with open('vectorizer.pkl','rb') as f:
    vectorizer = pickle.load(f)
    
```

```python
import pandas as pd
df_test = pd.read_csv("D://kaggle//test.csv")
df_new = {'0':[]}
c = 5
for i in range(0,595300):
    text = df_test[' Log'][i].split()[5:]
    text = ' '.join(text)
    df_new['0'].append(text)
print("Done")
```

```python
df_new = pd.DataFrame(df_new)
df_new
```

```python
c = 5
y_predict = []
x_test = df_new['0']
import numpy as np
# for i in range(595300):
#     text = df_new['0'][i]
#     text = vectorizer.transform([text])
#     text = text.toarray()
#     y_predict.append(model.predict(text))
x_test = np.array_split(x_test,10000)
```

```python
for i in range(0,10000):
    x = x_test[i]
    x = vectorizer.transform(x)
    x = x.toarray()
    pre = model.predict(x)
    y_predict.extend(pre)
    if (i+1) % 1000 == 0:
        print(i+1)
```

```python
dic = {0:'normal',1:'abnormal'}
for i in range(0,595300):
    y_predict[i] = dic[y_predict[i]]
y_predict
```
